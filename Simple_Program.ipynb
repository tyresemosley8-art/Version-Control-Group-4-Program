```python
# A simple program to say hello and do some math
name = input("Enter your name: ")
favorite_number = input("Enter your favorite number: ")

# We have to turn the input into a number so we can do math
number = int(favorite_number)
result = number * 10

print("Hello " + name + "!")
print("If you multiply your favorite number by 10, you get: " + str(result))
```
